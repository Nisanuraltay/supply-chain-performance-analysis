Veri temizlemeye giri≈ü

```python
# Lojistik analizi i√ßin gerekli kolonlar
columns_to_keep = [
    # Sipari≈ü bilgileri
    'Order Id',
    'order date (DateOrders)', # 'Order Date (DateOrders)' yerine d√ºzeltildi
    'Days for shipping (real)',
    'Days for shipment (scheduled)',
    'Delivery Status',

    # M√º≈üteri/Coƒürafya
    'Customer Country',
    'Customer City',
    'Customer State',
    'Order Country',
    'Order Region',

    # √úr√ºn bilgileri
    'Product Name',
    'Category Name',
    'Department Name',

    # Finansal
    'Order Item Quantity',
    'Sales',
    'Order Item Total',

    # Lojistik
    'Shipping Mode',
    'Late_delivery_risk',

    # Tarih
    'shipping date (DateOrders)'
]

# Sadece bu kolonlarƒ± al
df_clean = df[columns_to_keep].copy()

print(f"‚úÖ {len(columns_to_keep)} kolon se√ßildi")
print(f"üìä Yeni boyut: {df_clean.shape}")
```

Kolon isimlerini d√ºzenleme

```python
# Daha okunaklƒ± kolon isimleri
df_clean.columns = [
    'order_id',
    'order_date',
    'actual_shipping_days',
    'scheduled_shipping_days',
    'delivery_status',
    'customer_country',
    'customer_city',
    'customer_state',
    'order_country',
    'order_region',
    'product_name',
    'category',
    'department',
    'quantity',
    'sales',
    'order_total',
    'shipping_mode',
    'late_delivery_risk',
    'shipping_date'
]

print("‚úÖ Kolon isimleri d√ºzenlendi")
df_clean.head()
```

tarih kolonlarƒ±nƒ± d√ºzenleme

```python
# Tarih formatƒ±nƒ± d√ºzelt
df_clean['order_date'] = pd.to_datetime(df_clean['order_date'], errors='coerce')
df_clean['shipping_date'] = pd.to_datetime(df_clean['shipping_date'], errors='coerce')

# Yeni √∂zellikler t√ºret
df_clean['year'] = df_clean['order_date'].dt.year
df_clean['month'] = df_clean['order_date'].dt.month
df_clean['quarter'] = df_clean['order_date'].dt.quarter
df_clean['day_of_week'] = df_clean['order_date'].dt.dayofweek
df_clean['month_name'] = df_clean['order_date'].dt.strftime('%B')

print("‚úÖ Tarih √∂zellikleri eklendi")
print(f"\nüìÖ Tarih aralƒ±ƒüƒ±: {df_clean['order_date'].min()} - {df_clean['order_date'].max()}")
```

Gecikme metrikleri hesaplama

```python
# Gecikme hesaplama
df_clean['delay_days'] = df_clean['actual_shipping_days'] - df_clean['scheduled_shipping_days']
df_clean['is_delayed'] = (df_clean['delay_days'] > 0).astype(int)
df_clean['on_time'] = (df_clean['delivery_status'] == 'Shipping on time').astype(int)

# Gecikme kategorileri
df_clean['delay_category'] = pd.cut(
    df_clean['delay_days'],
    bins=[-np.inf, 0, 3, 7, np.inf],
    labels=['Zamanƒ±nda', '1-3 G√ºn', '4-7 G√ºn', '7+ G√ºn']
)

print("‚úÖ Gecikme metrikleri eklendi")
print(f"\n‚è±Ô∏è Gecikme ƒ∞statistikleri:")
print(df_clean['delay_days'].describe())
print(f"\nüìä Gecikmeli Sipari≈ü Oranƒ±: {df_clean['is_delayed'].mean():.2%}")
```

Eksik deƒüerleri y√∂netme

```python
# Eksik deƒüer analizi
print("üîç Eksik Deƒüer Kontrol√º:")
missing = df_clean.isnull().sum()
missing_pct = (missing / len(df_clean) * 100).round(2)
missing_df = pd.DataFrame({
    'Eksik Sayƒ±': missing,
    'Eksik (%)': missing_pct
})
print(missing_df[missing_df['Eksik Sayƒ±'] > 0].sort_values('Eksik Sayƒ±', ascending=False))

# Kritik kolonlardaki eksikleri √ßƒ±kar
df_clean = df_clean.dropna(subset=['order_date', 'customer_country', 'shipping_mode'])

# Kategorik deƒüi≈ükenlerde eksikleri 'Unknown' ile doldur
categorical_cols = ['customer_city', 'customer_state', 'category', 'department']
for col in categorical_cols:
    if col in df_clean.columns:
        df_clean[col] = df_clean[col].fillna('Unknown')

print(f"\n‚úÖ Temizleme sonrasƒ± boyut: {df_clean.shape}")
```

Aykƒ±rƒ± deƒüerleri kontrol etme

```python
# Negatif deƒüerleri kontrol et
print("üîç Negatif Deƒüer Kontrol√º:")
numeric_cols = ['quantity', 'sales', 'order_total', 'actual_shipping_days', 'scheduled_shipping_days']

for col in numeric_cols:
    neg_count = (df_clean[col] < 0).sum()
    if neg_count > 0:
        print(f"  ‚ö†Ô∏è {col}: {neg_count} negatif deƒüer bulundu")
        # Negatif deƒüerleri sil veya d√ºzelt
        df_clean = df_clean[df_clean[col] >= 0]

# A≈üƒ±rƒ± u√ß deƒüerleri kontrol et (IQR y√∂ntemi)
def remove_outliers(df, column, threshold=3):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - threshold * IQR
    upper = Q3 + threshold * IQR

    outliers = ((df[column] < lower) | (df[column] > upper)).sum()
    print(f"  üìä {column}: {outliers} aykƒ±rƒ± deƒüer")

    return df[(df[column] >= lower) & (df[column] <= upper)]

# Sadece gecikme i√ßin aykƒ±rƒ± deƒüerleri √ßƒ±kar
print("\nüéØ Aykƒ±rƒ± Deƒüer Analizi:")
df_clean = remove_outliers(df_clean, 'delay_days', threshold=3)

print(f"\n‚úÖ Final boyut: {df_clean.shape}")
```
